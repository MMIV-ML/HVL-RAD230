Det er to neste viktige operasjoner mellom vektorer: vektoriske og skalare multiplikasjoner. <br>
Det er viktig å forstå at for skalare multiplikasjonen vi får et tall. <br>
Når for vektorisk operasjonen får vi en vektor igjen.<br>
<br>  
Skalare multipliksjonen er: $v\cdot w = a$, hvor $a$ er et tall.

```julia
using LinearAlgebra # denne linjen gir oss ekstra programming libs med funksjoner
w = [5,4,3,2,1] # en mer vektor
v⋅w # skalare multiplikasjonen
```

hva det betyr? Vi kan få det samme resultatet hvis bruker element ved element operasjonen:<br>
$v\cdot w = (v_1,...,v_5)\cdot (w_1,...,w_5) = v_1w_1 + ... + v_5 w_5 = \sum_i v_i\cdot w_i$

```julia
w[1]*v[1]+w[2]*v[2]+w[3]*v[3]+w[4]*v[4]+w[5]*v[5]
```

Da kan vi snakke litt om matriser og hva slags matriser kan man tenke på?<br>
<br>
symmetrisk:<br>
    $S = \left[ \begin{array}{cc} 1&2 \\ 2&3 \end{array} \right]$<br>
diagonal:<br>
    $D = \left[ \begin{array}{cc} 1&0 \\ 0&2 \end{array} \right]$<br>
identitetsmatrise<br>
    $I = \left[ \begin{array}{cc} 1&0 \\ 0&1 \end{array} \right]$<br>

```julia
A = [2 15; 8 45];
S = [1 2; 2 3];
D = [1 0; 0 2];
IM = [1 0; 0 1];
A*S
```

```julia
S*A
```

Man kan se her at $S\cdot A$ er ikke det samme som $A\cdot S$. Dette trekket er kjent som non-kommutative matriser.<br>
Hvis $SA-AS = 0$ da det er kommutativ matriser. Det er veldig viktige resultatet fordi det bringer mange viktige konsekvenser for matimatikk.

Andre ting om matriser er eigenverdier, i.e. noen verdier hvilke er uavhengig av matrise formen eller matrise i eigenrom er altid diagonal.

```julia
B = eigvals(A)
```

```julia
C = eigvecs(A)
```

Man kan forstå det som noen notasjon ned:<br>
$Au = \lambda u$, hvor A er matrisen vår, $u$ er eigenvektor og $\lambda$ er eigenverdi.<br>
Et viktig resultat er: noen symmetriske matrise kan bli representert som diagonal og eigenvektor.<br>
Faktisk betyr det følgende:

```julia
A = [1 2 3; 2 4 5; 3 5 6]
```

```julia
Al = eigvals(A)
```

```julia
Au = eigvecs(A)
```

```julia
Au'*A*Au
```

# Hvorfor er viktig å jobbe med matriser og vektorer?

Vi kan manipulere med vectorer og matriser når vi bruker spesielle matriser. <br>
For eksempel, rotasjon matrise (også vi vet den som orthogonal matrise eller <br>
$R\cdot R^T = R^T\cdot R = I$ identitetsmatrisen.)<br>
Rotasjon matrise for 2D tillfelet: <br>
    $R = \left( \begin{array}{cc} \cos(\theta) & \sin(\theta)\\ -\sin(\theta) & \cos(\theta) \end{array} \right)$

```julia
θ = π/4
R = [cos(θ) sin(θ); -sin(θ) cos(θ)]
```

```julia
v1 = [0; 1]
plot([0,v1[1]],[0,v1[2]],
    label = L"$v$",
    arrow = :head,
    linewidth = 5)
```

```julia
v2 = R*v1
plot([0,v2[1]],[0,v2[2]],
    label = L"$v$",
    arrow = :head,
    linewidth = 5)
```

```julia
v3 = round.(R*v2)
plot([0,v3[1]],[0,v3[2]],
    label = L"$v$",
    arrow = :head,
    linewidth = 5)
```

# Oppløsningen er et eksampel av smoothing teknikk.

```julia
using NaNStatistics
N = 1000
x = 1:N
θ = x*2π/N
tr = sin.(θ)
y = sin.(θ) + rand(N,1) .-0.5
plot(θ,y)
plot!(θ,tr, linewidth = 5, color = :orange)
z = movmean(y,N/100)
plot!(θ,z,linewidth = 2, color = :magenta)
```

# Smoothing teknikk er ganske enkel

```julia
a = [5.1, 5.2, 4, 4.2, 5.1, 4.9, 5.2]
plot(a, linewidth = 4)
function smoothing(a)
    N = length(a);
    b = zeros(N-2,1);
    for i = 1:(N-2)
        b[i] =  (a[i] + a[i+1]+a[i+2])/3;
    end
    return b
end
b = smoothing(a)
plot!([1.5, 2.5, 3.5, 4.5, 5.5],b,linewidth = 3, ylim = [3.5,5.5])
```
